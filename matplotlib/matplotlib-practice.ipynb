```python
walmart_df = pd.read_csv('..//zero_analyst//30_days_python_bk//day20_walmart.csv')
# walmart_df
city_average_price  = walmart_df.groupby(by='city').agg({ 'unit_price' :'mean'}).reset_index()
city_average_price['unit_price'] = city_average_price['unit_price'].round(2)
# print(city_average_price)
```

```python
# city_average_price.loc[:]
city_average_price['unit_price']
```

```python
# walmart city average unit_price plotting using vertical  bar 

plt.figure(figsize=( 12.0, 4.0))
plt.bar(city_average_price['city'] , city_average_price['unit_price']  , width=[0.4 , ] , color='#81BFDA')
# plt.bar(city_average_price['city'] , city_average_price['unit_price']  , width=[0.6 , ] , color='#81BFDA')
plt.title(label='Walmart city Avg unit_price Chart')
plt.xlabel('City')
plt.ylabel('Unit Price in $' )
plt.grid(axis='y',linestyle='--' ,color = '#355F2E')
# plt.legend()
plt.show()
```

```python
# walmart city average unit_price plotting using horizental bar 

plt.figure(figsize=( 8.0, 4.0) )
plt.rcParams['font.family'] = 'Comic Sans MS'
plt.barh(city_average_price['city'] , city_average_price['unit_price']  , height=[0.4 , ] , color='#FCC737' )
plt.title(label='Walmart city Avg unit_price Chart')
plt.ylabel('City')
plt.xlabel('Unit Price' )
plt.grid(axis='x',linestyle='--' ,color = '#355F2E')
# plt.legend()
plt.show()
```

```python
# walmart city average unit_price plotting using simple plot 
plt.figure(figsize=( 12.0, 4.0) )
plt.rcParams['font.family'] = 'Comic Sans MS'
plt.plot(city_average_price['city'] , city_average_price['unit_price'] , marker='o'   , color='#FCC737' )
plt.title(label='Walmart city Avg unit_price Chart')
plt.ylabel('City')
plt.xlabel('Unit Price' )
# plt.grid(axis='x',linestyle='--' ,color = '#355F2E')
# plt.legend()
plt.show()
```

```python
walamart_gender_dist = walmart_df.groupby(by='branch').agg({'invoice_id': 'count'}).reset_index().rename(columns={'invoice_id' :'counts'})
walamart_gender_dist
```

```python
# Walmart Branch Distribution Chart 
plt.figure(figsize=( 8.0, 4.0))
plt.rcParams['font.family'] = 'Comic Sans MS'
explodes  = np.ones(5)*0.0
explodes[3] = 0.1
colors = np.random.randint(10,10000 , 5)
colors=np.sort(colors)
colors = colors/ colors[0]
norm_colors = plt.cm.viridis(colors)
plt.pie( walamart_gender_dist['counts']  , explode=explodes , colors=norm_colors ,   labels=walamart_gender_dist['branch'] , autopct='%1.2f%%'  , shadow=True , ) 
plt.title(label='Walmart Branch Distribution Chart')
plt.show()
```
