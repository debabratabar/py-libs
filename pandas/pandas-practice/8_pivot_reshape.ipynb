## We can reshape dataframe using 
- pivot(index , columns , values )
- pivot_table(index , columns , values , aggFunc)
- melt()
- stacking ( col to row)  & unstacking ( row to column )
- crosstab

```python
import pandas as pd 
```

```python
weather = pd.read_csv('weather.csv')
weather
```

```python
## using pivot we can reshape the dataframe 
df1 = pd.pivot(weather , index='date' , columns='city' )
df1
```

```python
# just checking how the data looks in excel format 
df1.to_excel('weather.xlsx'  , sheet_name='sheet1')
```

```python
weather3 = pd.read_csv('weather3.csv')
weather3
```

```python
df2 = weather3.pivot(index='city' , columns='date' , values='humidity')
df2 
```

```python
weather2 = pd.read_csv('weather2.csv')
weather2
```

```python
max_temp = weather2.pivot_table(index='city' , columns='date' , values='temperature' , aggfunc='max' )
max_temp
```

```python
weather = pd.read_csv('weather.csv')
weather
```

```python
# using melt we are creating variable & value data for every id_vars 
df1 = pd.melt(weather , id_vars=['city'])
df1[df1['variable']=='date']
```
