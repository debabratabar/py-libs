```python
import pandas as pd 
```

```python
stock_data = pd.read_csv('stock_data.csv')
stock_data
```

```python
stock_data = pd.read_csv('stock_data.csv'  , nrows=3) ## number of rows you want to read 
stock_data
```

```python
# using skiprows to skip 1 row ( header ) and giving user-defined header or column-name 
stock_data = pd.read_csv('stock_data.csv' , skiprows=1 , names=['tickers' , 'eps' , 'revenue' , 'price' , 'people '])
stock_data
```

```python
# remove all the NA values to NaN 
stock_data = pd.read_csv('stock_data.csv' , na_values=['not available' , 'n.a.']) # for common use 
# stock_data

# if you have multiple cond. for NA values for every col , we can use a dict 
stock_data = pd.read_csv('stock_data.csv' , na_values={
      'eps':['not available' , 'n.a.'],
      'people' : ['not available' , 'n.a.'],
    'revenue' :['not available' , 'n.a.',-1],
    'price' : ['not available' , 'n.a.'],
}
    ) 
stock_data
```

```python
stock_data.to_csv('tr1.csv' ,columns=['tickers' , 'price']  , index=False)
```

```python
stock_data.to_csv('tr2.csv' , doublequote=True , index=False)
```
