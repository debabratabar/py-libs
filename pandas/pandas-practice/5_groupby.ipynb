## using group by 

```python
import pandas as pd
```

```python
weather_data = pd.read_csv('weather_by_cities.csv')
weather_data
```

```python
weather_grp_by = weather_data.groupby(['city']).max()
weather_grp_by[['temperature']]
```

```python
for city , city_df  in weather_grp_by:
    print(city)
    print(city_df)
```

```python
weather_grp_by = weather_data.groupby(['city'])
weather_grp_by.describe()
```

```python
# by default dropna is True , if there is any Na value in key col.  then it will be dropped
weather_grp_by = weather_data.groupby('city' ,  dropna=True  ).max()
weather_grp_by
```

```python
# get_group to get the individual key data 
weather_grp_by = weather_data.groupby(['city'])
weather_grp_by.get_group('paris')
```
