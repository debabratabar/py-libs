```python
import pandas as pd
```

```python
india_weather = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
    "humidity": [80, 60, 78]
})
india_weather
```

```python
us_weather = pd.DataFrame({
    "city": ["new york","chicago","ohio"],
    "temperature": [32,45,30],
    "humidity": [80, 60, 78]
})
us_weather
```

```python
df = pd.concat([india_weather , us_weather] )
df
```

```python
df = pd.concat([india_weather , us_weather]  ,  keys=['ind' , 'usa'] )
df
```

```python
temp_df = pd.DataFrame({
    "city": ["new york","chicago","ohio"],
    "temperature": [32,45,30],
}, index=[0,1,2])
temp_df

humidity_df = pd.DataFrame({
    "city": ["chicago","new_york","ohio"],
    "humidity": [80, 60, 78]
}, index=[1,0,2]  )
humidity_df
```

```python
new_df  = pd.concat( [ temp_df , humidity_df ] , axis=1)
new_df
```
